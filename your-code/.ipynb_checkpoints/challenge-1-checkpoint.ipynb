#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.figure()
#(rows,columns,index)
#Esto es el primero
plt.subplot(1,2,1) 
plt.plot(x,y,color='m',linewidth=2,linestyle='-',marker='o',label='Uno')
plt.title('Ejercicio 1')
plt.legend()

#Esto es el segundo
plt.subplot(1,2,2)
plt.plot(y,z,color='b',linewidth=3,linestyle=':',marker=',',label='Dos')
plt.title('Ejercicio 2')
plt.legend()

#Terminar gráfica
plt.show()
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x,y,color='m',linewidth=2,linestyle='-',marker='o',label='Uno')
ax1.set_title('Ejercicio 1')
ax1.legend()

ax2.plot(y,z,color='b',linewidth=3,linestyle=':',marker=',',label='Dos')
ax2.set_title('Ejercicio 2')
ax2.legend()

plt.show()
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, [ax1, ax2] = plt.subplots(1,2,figsize=(5,7)) #(weight,height)
ax1.plot(x,y,color='m',linewidth=2,linestyle='-',marker='o',label='Uno')
ax1.set_title('Ejercicio 1')
ax1.legend()
ax1.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')


ax2.plot(y,z,color='b',linewidth=3,linestyle=':',marker=',',label='Dos')
ax2.set_title('Ejercicio 2')
ax2.legend()
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')

plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
x = np.arange(0,100)
y1 = x**2
y2 = np.exp(x)

plt.figure()
#(rows,columns,index)
#Esto es el primero
plt.subplot(1,2,1) 
plt.plot(x,y1,color='r',linewidth=1)
plt.plot(x,y2,color='r',linewidth=1)
plt.title('Ejercicio 1')
plt.ylabel('y')
plt.xlabel('x')
plt.xscale('linear')
plt.yscale('linear')


plt.subplot(1,2,2) 
plt.plot(x,y1,color='g',linewidth=1)
plt.plot(x,y2,color='r',linewidth=1)
plt.title('Ejercicio 2')
plt.ylabel('y')
plt.xlabel('x')
plt.xscale('log')
plt.yscale('log')

```
